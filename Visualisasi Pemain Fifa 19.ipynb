```python
kordata = ['Age', 'Stamina' , 'Strength' , 'Acceleration','SprintSpeed' , 'Agility' , 'Jumping' ,
       'Vision','Reactions']
```

```python
sns.heatmap(data[kordata].corr(),linewidths=.5,annot=True,cmap='RdYlGn')
fig=plt.gcf()
fig.set_size_inches(15,8)
plt.show()
```

```python
best_dict = {}
for club in data['Club'].unique():
    overall_rating = data['Overall'][data['Club'] == club].sum()
    best_dict[club] = overall_rating
```

```python
df_bcp = pd.DataFrame.from_dict(best_dict , orient = 'index')
```

```python
df_bcp.rename(columns={list(df_bcp)[0]:'skor'}, inplace = True)
df_bcp.head(10)
```

```python
df_bcp['skor'].nlargest(20).plot(kind = 'bar')
plt.show()
```
